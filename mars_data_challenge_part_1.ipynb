```python
# Import dependencies

from splinter import Browser
from bs4 import BeautifulSoup
from webdriver_manager.chrome import ChromeDriverManager

```

```python
# Setup splinter

executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# Open website

url = 'https://redplanetscience.com/'
browser.visit(url)
```

```python
 # HTML object
html = browser.html
 # Parse HTML with Beautiful Soup
soup = BeautifulSoup(html, 'html.parser')
# Print soup to get the element(s) that contain title and preview
soup
```

```python
# Select the element that contains title and preview
mars_news= soup.select('div.list_text')
mars_news
```

```python
# Create an empty list

news_list = []
```

```python
# Iterate through pages

for x in mars_news:
     
    title = x.select_one('div.content_title').get_text()
    preview = x.select_one('div.article_teaser_body').get_text()
    
    news_dict = {}
    
    news_dict["title"] = title
    news_dict["preview"] = preview
    
    news_list.append(news_dict)
    
news_list
 
```

```python
browser.quit()
```

```python
# Import dependencies

import json
```

```python
# Create a json file to store the scraped information and save as file

with open("mars_news_summaries.json", "w") as file:
    json.dump(news_list, file)
```
